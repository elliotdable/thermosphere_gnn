```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Add month and hour columns
full_fpi_df['month'] = full_fpi_df['datetime'].dt.month
full_fpi_df['hour'] = full_fpi_df['datetime'].dt.hour

# Define your custom month/hour orders
month_order = [8, 9, 10, 11, 12, 1, 2, 3, 4]
hour_order = list(range(13, 24)) + list(range(0, 8))

# ---- Monthly aggregates ----
monthly_stats = (
    full_fpi_df
    .groupby(['site', 'month'])
    .agg({'temperature': ['mean', 'std'], 'wind_speed': ['mean', 'std']})
    .reset_index()
)
monthly_stats.columns = ['site', 'month', 'temp_mean', 'temp_std', 'wind_mean', 'wind_std']

monthly_stats = monthly_stats[monthly_stats['month'].isin(month_order)]
monthly_stats['month'] = pd.Categorical(monthly_stats['month'], categories=month_order, ordered=True)
monthly_stats = monthly_stats.sort_values('month')
monthly_stats['month_name'] = monthly_stats['month'].apply(lambda m: calendar.month_abbr[m])

# ---- Hourly aggregates ----
hourly_stats = (
    full_fpi_df
    .groupby(['site', 'hour'])
    .agg({'temperature': ['mean', 'std'], 'wind_speed': ['mean', 'std']})
    .reset_index()
)
hourly_stats.columns = ['site', 'hour', 'temp_mean', 'temp_std', 'wind_mean', 'wind_std']

hourly_stats = hourly_stats[hourly_stats['hour'].isin(hour_order)]
hourly_stats['hour'] = pd.Categorical(hourly_stats['hour'], categories=hour_order, ordered=True)
hourly_stats = hourly_stats.sort_values('hour')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# ---- Temperature by Month ----
for site in monthly_stats['site'].unique():
    sdata = monthly_stats[monthly_stats['site'] == site]
    x = sdata['month_name']
    axes[0, 0].plot(x, sdata['temp_mean'], label=site)
    axes[0, 0].fill_between(x, sdata['temp_mean'] - sdata['temp_std'], 
                               sdata['temp_mean'] + sdata['temp_std'], 
                               alpha=0.15)
axes[0, 0].set_title('Temperature by Month (AugApr)')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Temperature (C)')
axes[0, 0].legend()
axes[0, 0].grid(True)

# ---- Wind Speed by Month ----
for site in monthly_stats['site'].unique():
    sdata = monthly_stats[monthly_stats['site'] == site]
    x = sdata['month_name']
    axes[1, 0].plot(x, sdata['wind_mean'], label=site)
    axes[1, 0].fill_between(x, sdata['wind_mean'] - sdata['wind_std'], 
                               sdata['wind_mean'] + sdata['wind_std'], 
                               alpha=0.15)
axes[1, 0].set_title('Wind Speed by Month (AugApr)')
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Wind Speed (m/s)')
axes[1, 0].legend()
axes[1, 0].grid(True)

# ---- Temperature by Hour ----
for site in hourly_stats['site'].unique():
    sdata = hourly_stats[hourly_stats['site'] == site]
    x = sdata['hour'].astype(str)
    axes[0, 1].plot(x, sdata['temp_mean'], label=site)
    axes[0, 1].fill_between(x, sdata['temp_mean'] - sdata['temp_std'], 
                               sdata['temp_mean'] + sdata['temp_std'], 
                               alpha=0.15)
axes[0, 1].set_title('Temperature by Hour (13:0007:00)')
axes[0, 1].set_xlabel('Hour')
axes[0, 1].set_ylabel('Temperature (C)')
axes[0, 1].legend()
axes[0, 1].grid(True)

# ---- Wind Speed by Hour ----
for site in hourly_stats['site'].unique():
    sdata = hourly_stats[hourly_stats['site'] == site]
    x = sdata['hour'].astype(str)
    axes[1, 1].plot(x, sdata['wind_mean'], label=site)
    axes[1, 1].fill_between(x, sdata['wind_mean'] - sdata['wind_std'], 
                               sdata['wind_mean'] + sdata['wind_std'], 
                               alpha=0.15)
axes[1, 1].set_title('Wind Speed by Hour (13:0007:00)')
axes[1, 1].set_xlabel('Hour')
axes[1, 1].set_ylabel('Wind Speed (m/s)')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import os
import ast
import torch
import numpy as np
import pandas as pd
from src.GNN.gnn_model import SpatioTemporalGCN
from src.GNN.load_data import run_msis_specific_location
from src.GNN.data_processing import inference_data_object

imf_path = '/Users/elliotdable/Documents/PhD/research/fpi_ml/data/imf_data/compiled/omni_imf_data_1990_2025_cleaned.csv'
geomag_path = '/Users/elliotdable/Documents/PhD/research/fpi_ml/data/geomag_data/solar_geomag_dst_data.csv'

def load_training_row(slurm_job_id, csv_path):
    """Load metadata and feature lists for a given SLURM job."""
    df = pd.read_csv(csv_path)

    required_cols = {
        'slurm_job_id', 'imf_feature_cols', 'geomag_feature_cols',
        'period', 'imf_hours', 'geomag_days', 'imf_resample', 'model_name',
        'dropout_hidden', 'dropout_in', 'hidden_dim'
    }
    if not required_cols.issubset(df.columns):
        missing = required_cols - set(df.columns)
        raise ValueError(f"CSV is missing required columns: {missing}")

    row = df.loc[df['slurm_job_id'] == slurm_job_id]
    if row.empty:
        raise ValueError(f"No row found for slurm_job_id = {slurm_job_id}")

    row = row.iloc[0]

    model_path = os.path.join(
        '/Users/elliotdable/Documents/PhD/research/fpi_ml/models',
        row['model_name']
    )

    return {
        "imf_features": ast.literal_eval(row['imf_feature_cols']),
        "geomag_features": ast.literal_eval(row['geomag_feature_cols']),
        "period": row['period'],
        "imf_hours": row['imf_hours'],
        "geomag_days": row['geomag_days'],
        "imf_resample": row['imf_resample'],
        "model_path": model_path,
        "dropout_hidden": row['dropout_hidden'],
        "dropout_in": row['dropout_in'],
        "hidden_dim": row['hidden_dim'],
    }


def load_model_and_predict(data_object, model_path, hidden_dim, dropout_in, dropout_hidden, device):
    """Load the trained GNN model and run inference."""
    checkpoint_path = f"{model_path}/model_best.pth"
    y_scaler = torch.load(f"{model_path}/target_scaler.pt")

    in_channels = data_object.x.shape[1]
    edge_attr_dim = data_object.edge_attr.shape[1]

    model = SpatioTemporalGCN(
        in_channels=in_channels,
        edge_attr_dim=edge_attr_dim,
        hidden_channels=int(hidden_dim),
        dropout_in=dropout_in,
        dropout_hidden=dropout_hidden
    ).to(device)

    # Load checkpoint (remove compiled wrapper keys)
    state_dict = torch.load(checkpoint_path, map_location=device)
    cleaned_state = {k.replace("_orig_mod.", ""): v for k, v in state_dict.items()}
    model.load_state_dict(cleaned_state)
    model.eval()

    with torch.inference_mode():
        data = data_object.to(device)
        out = model(data.x, data.edge_index, data.edge_attr)

    pred = out.detach().cpu().numpy().reshape(-1, 1)
    pred_phys = y_scaler.inverse_transform(pred)

    return pred_phys


def run_storm_inference(slurm_job_id, storm_df, imf_path, geomag_path, training_csv_path):
    """
    Complete pipeline for inference:
    - loads metadata using slurm_job_id
    - creates inference DataObject
    - loads GNN model and predicts
    - computes RMSEs
    - computes MSIS temperatures
    - returns prediction dataframe
    """
    # Load training info
    meta = load_training_row(slurm_job_id, training_csv_path)

    # Build inference data object
    data_object = inference_data_object(
        fpi_df=storm_df,
        imf_path=imf_path,
        geomag_path=geomag_path,
        use_pca=meta['use_pca'] if 'use_pca' in meta else False,
        start_date=storm_df['datetime'].min(),
        end_date=storm_df['datetime'].max(),
        imf_feature_cols=meta["imf_features"],
        imf_hours=meta["imf_hours"],
        geomag_feature_cols=meta["geomag_features"],
        model_path=meta["model_path"],
        imf_resample=meta["imf_resample"],
        geomag_days=meta["geomag_days"]
    )

    # Run inference
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    pred_phys = load_model_and_predict(
        data_object=data_object,
        model_path=meta["model_path"],
        hidden_dim=meta["hidden_dim"],
        dropout_in=meta["dropout_in"],
        dropout_hidden=meta["dropout_hidden"],
        device=device
    )

    # RMSE
    rmse_phys = np.sqrt(np.mean((pred_phys - storm_df['temperature'].to_numpy()) ** 2))
    print(f"RMSE (physical units): {rmse_phys:.4f} K")

    # Add predictions + MSIS
    predictions = storm_df.copy()
    predictions["pred_phys"] = pred_phys

    predictions = predictions.rename(columns={
        'observation_latitude': 'geodetic_latitude',
        'observation_longitude': 'geodetic_longitude',
        'altitude': 'geodetic_altitude'
    })

    msis_vals = run_msis_specific_location(predictions, version=2.1)
    predictions["msis_temperature"] = msis_vals["msis_temperature"].to_list()

    return predictions
```
